```python
import pandas as pd
import numpy as np
df=pd.read_csv('indian_liver_patient.csv')
df.head(10)
```

```python
df.isnull().sum()
```

```python
df2=df.dropna()
df2
```

```python
df2.isnull().sum()
```

```python
df2.dtypes
```

```python
df2['Gender']
df2
```

```python
d1=pd.get_dummies(df2['Gender'])
d1
```

```python
#merging
df3=pd.concat([d1,df2],1)
df3
```

```python
#dropping
df4=df3.drop(['Gender','Female'],axis=1)
df4
```

```python
X=df4.drop(['Dataset','Total_Protiens','Albumin_and_Globulin_Ratio','Albumin'],axis=1)
X
```

```python
y=df4['Dataset']
y
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
select=SelectKBest(score_func=chi2,k='all')
fit=select.fit(X,y)
```

```python
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
```

```python
featureScores = pd.concat([dfscores,dfcolumns],axis=1)
featureScores
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=0)
```

```python
from sklearn.svm import SVC
model2=SVC(kernel='linear')
model2
```

```python
model2.fit(X_train,y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

```python
y_pred=model2.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```
