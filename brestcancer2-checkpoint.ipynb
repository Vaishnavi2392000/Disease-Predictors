```python
import pandas as pd
import numpy as np
df=pd.read_csv('wisc_bc_data.csv')
df.head(10)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['diagnosis'].replace({'M':0,'B':1},inplace=True)
df
```

```python
df.dtypes
```

```python
X=df.drop(['smoothness_mean','symmetry_mean',
           'texture_se','smoothness_se',
           'compactness_se','concave points_se',
           'symmetry_se','smoothness_worst',
           'fractal_dimension_se','diagnosis','id','fractal_dimension_worst'],axis=1)
X
```

```python
y=df['diagnosis']
y
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
select=SelectKBest(score_func=chi2,k='all')
select
```

```python
fit=select.fit(X,y)
```

```python
dfscores=pd.DataFrame(fit.scores_)
dfcols=pd.DataFrame(X.columns)
```

```python
features=pd.concat([dfscores,dfcols],axis=1)
features
```

```python
features.columns=['specs','value']
```
