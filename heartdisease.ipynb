```python
import pandas as pd
df=pd.read_csv('datasets_4123_6408_framingham.csv')
df.head(10)
```

```python
#check_NAN_Values
df.isnull().sum()
```

```python
df1=df.drop(['education'],axis=1)
df1
```

df1.isnull().sum()

```python
df1.isnull().sum()
```

```python
#conerting NAN values to Non-NAN values
df2=df1.interpolate()
df2
```

```python
df2.isnull().sum()
```

```python
#variable_Declarations
X=df2.iloc[:,:14]
X
```

```python
y=df2['TenYearCHD']
y
```

```python
#spliting
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression(max_iter=1200000000)
```

```python
reg
```

```python
reg.fit(X_train,y_train)
```

```python
print(y_test)
```

```python
print(X_test)
```

```python
y_pred=reg.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_pred,y_test)
cm
```

```python
from sklearn.model_selection import cross_val_score
score=cross_val_score(reg,X,y,cv=10)
score
```

```python
score.mean()
```
